```python
import pandas as pd
import numpy
import matplotlib as plt
import pickle as pkl
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv("data/comic_characters.csv")
pd.DataFrame(df)
```

```python
eye = {"Blue": 0, "Brown": 1, "Green": 2, "Black": 3, "White": 4, "Variable": 5, "Hazel": 6, "Yellow": 7, "Red": 8, "Grey": 9, "Purple": 10, "Orange": 11, "Aubrun": 12, "Photocellular": 13, "Silver": 14}
df["Eyes"] = df["Eyes"].map(eye)

hair = {"Black": 0, "Red": 1, "White": 2, "Brown": 3, "Blond": 4, "No": 5, "Bald": 6, "Green": 7, "Orange": 8, "Variable": 9, "Grey": 10, "Blue": 11, "Auburn": 12, "Silver": 13, "Yellow": 14}
df["Hair"] = df["Hair"].map(hair)

i = {"Secret": 0, "Public": 1}
df["Identity"] = df["Identity"].map(i)

al = {"Good": 0, "Bad": 1}
df["Alignment"] = df["Alignment"].map(al)

sex = {"Male": 0, "Female": 1}
df["Sex"] = df["Sex"].map(sex)

alive = {"Yes": 0, "No": 1}
df["Alive"] = df["Alive"].map(alive)

Uni = {"DC": 0, "Marvel": 1}
df["Universe"] = df["Universe"].map(Uni)

features = ["Eyes", "Hair", "Identity", "Alignment", "Sex", "Alive", "Universe"]

X = df[features]
y = df["Name"]
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(X, y)

model_filename = 'MarvelDCName.pkl'
pkl.dump(dtree, open(model_filename,'wb'))
```

```python
model = pkl.load(open('MarvelDCName.pkl','rb'))
print(model.predict([[13, 1, 1, 1, 0, 0, 1]]))
```
